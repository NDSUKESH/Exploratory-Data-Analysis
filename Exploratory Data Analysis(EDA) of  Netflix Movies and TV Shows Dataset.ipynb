```python
ax=sns.countplot(x='type',data=df)
# Calculate the total count of each category
total = len(df['type'])

# Iterate over the bars and annotate the count and percentage
for p in ax.patches:
    count = p.get_height()
    percentage = f'{(count/total) * 100:.2f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()

    ax.annotate(f'{count}\n{percentage}', (x, y), ha='center', va='bottom')

# Set labels and title
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Countplot with Percentage')
```

- The Director column has more null value and that column is not more important in analysis so i decided to drop that column
- the same for cast column

```python
df.drop('director',axis=1,inplace=True)
df.drop('cast',axis=1,inplace=True)
df.isnull().sum()
```

```python
# there is only 11 NaN values in Date_added so i just scraped it
df = df[df['date_added'].notna()]
```

- We can see that both in Tv Shows and Movies the US is pretty much high compared to others so we caan use us to replace the missing values

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['year_added'] = df['date_added'].apply(lambda x: x.split(" ")[-1])

# spliting the month from the date_added column
df['month_added'] = df['date_added'].apply(lambda x: x.split(" ")[0])
```

```python
df.head()
```

```python
#we dont need that column
df.drop('date_added',axis=1,inplace=True)
```

```python
# Creating a dict to find the most targeted audience in Netflix show
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
df['target_ages'] = df['rating'].replace(ratings_ages)
df['target_ages'].unique()
```

```python
sns.countplot(x='target_ages',data=df)
```

Most of the show in Netflix is adult targeted shows followed by Teenage

```python
sns.countplot(x='target_ages',data=df,hue='type',palette='rainbow')
```
